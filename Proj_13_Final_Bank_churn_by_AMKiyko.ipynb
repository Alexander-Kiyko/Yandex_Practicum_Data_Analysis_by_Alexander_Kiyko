Сформируем соответствующий датафрейм и посмотрим.

```python
# задаем датафрейм с перечисленными параметрами
df_churn_loyal = pd.DataFrame(
        index=['количество клиентов',
               'наиболее частый город',
               'наиболее частый пол',
               'средний возраст',
               'средняя зарплата',
               'медианный скоринг',
               'мед. кол-во объектов собственности',
               'среднее количество продуктов банка',
               'медианный остаток на счете',
               'доля активных пользователей',
               'доля владельцев кредитных карт'
              ],
        columns=['Лояльные','Отточные']
)

# заполняем датафрейм значениями
for i, col in enumerate(df_churn_loyal.columns):
    df_churn_loyal.loc['количество клиентов', col] = len(df[df.churn == i])
    df_churn_loyal.loc['наиболее частый город', col] = df[df.churn == i]['city'].mode()[0]
    df_churn_loyal.loc['наиболее частый пол', col] = df[df.churn == i]['gender'].mode()[0]
    df_churn_loyal.loc['средний возраст', col] = round(df[df.churn == i]['age'].mean(), 0)
    df_churn_loyal.loc['средняя зарплата', col] = round(df[df.churn == i]['est_salary'].mean(), 0)
    df_churn_loyal.loc['медианный скоринг', col] = round(df[df.churn == i]['score'].median(), 2)
    df_churn_loyal.loc['мед. кол-во объектов собственности', col] = round(df[df.churn == i]['equity'].median(), 0)
    df_churn_loyal.loc['среднее количество продуктов банка', col] = round(df[df.churn == i]['products'].mean(), 0)
    df_churn_loyal.loc['медианный остаток на счете', col] = round(df[df.churn == i]['balance'].median(), 0)
    df_churn_loyal.loc['доля активных пользователей', col] = \
                        round(len(df[(df.churn == i) & ((df.activity == 1))]) / len(df[df.churn == i]), 2)
    df_churn_loyal.loc['доля владельцев кредитных карт', col] = \
                        round(len(df[(df.churn == i) & ((df.credit_card == 1))]) / len(df[df.churn == i]), 2)

df_churn_loyal
```

### Корреляционный анализ признаков <a id='step_3.4'></a>

Посмотрим взаимозависимости признаков. Для этого построим матрицц линейной корреляции Пирсона. Однако есть подозрение, что линейная корреляция в данном случае может не дать уверенного результата. Поэтому если показтели будут невнятные, то учтем нелинейные зависимости, построив матрицу корреляции с помощью библиотеки phik. 

Перед корреляционным анализом приведем текстовые столбцв датафрейма к числовому виду.  Таких столбцов у нас два: "age" и "city".                
Столбец "age" по сути булев. Поэтому наиболее логично преобразовать его задав значению Ж числовой параметр 0 и значению М - парметр 1.          
А для столбца "city" проделаем привычную в таких случаях процедуру pd.get_dummies(df).    
Кроме того нужно бы удалить столбец "user_id" - он для корреляционного анализа не нужен.

```python
# формируем датафрейм для проведения корреляционного анализа
df_corr = df

# преобразуем данные столбца "gender"
df_corr.replace(['Ж','М'],[0,1], inplace=True)

# преобразуем данные столбца "city" 
df_corr = pd.get_dummies(df)

# удалим столбец "user_id"
df_corr = df_corr.drop('user_id', axis=1)

# смотрим что получилось
df_corr.head()
```

Кроме того ранее мы отмечали определенные "группы риска" с точки зрения оттока в категориях "возраст" и "уровень кредитного скоринга". Здесь же мы видим, что показатели корреляции оттока с этими параметрами невысоки.           
Хорошо бы хотя бы некоторые из этих параметров проверить на статистическую значимость.

## Шаг 4. Проверка статистических гипотез по отдельным параметрам  <a id='step_4'></a>

Как мы увидели ранее у нас есть доволно важные параметры, влияние которых на отток неочевидно.
Прежде всего это возраст клиента, оценка его заработной платы и баланс на счете. 
Ну и наверное будет не лишним проверить еще и кредитный скоринг.                  

Сформулируем гипотезы о равенстве этих параметров для отточных клиентов и лояльных клиентов и попробуем проверить можем ли мы статистически значимо принять или отвергнуть такую гипотезу.           

Как мы отмечали ранее все три эти параметра имеют явную скошенность распределения вправо и довольно много аномально высоких значений. 
Поэтому для проверки статистической значимости гипотезы в этом случае лучше всего будет использовать  тест Уилкоксона-Манна-Уитни, который, как нас учили, устойчив к выбросам. 
Однако все-таки кажды раз мы будем проверять нормальность распределения по методу Шапиро-Уилка (st.shapiro) и если вдруг проверка покажет, что распределение таки нормальное - то будем использовать для проверки проверки гипотез t-критерий, применяемый при нормальном распределении.

 ### Проверка гипотезы для параметра возраст <a id='step_4.1'></a>

```python
print("Средний возраст отточного клиента - ", df.query('churn == 1')['age'].mean().round(1))
print("Средний возраст лояльного клиента - ", df.query('churn == 0')['age'].mean().round(1))
```

Проверим нормальность распределения методом Шапиро-Уилка (st.shapiro)           
с нулевой гипотезой Н0 - что распределение параметра нормально,                
и альтернативной гипотезой Н1 - что распределение не является нормальным.

```python
# зададим данные для анализа
group = df['age']

# задаем критический уровень статистической значимости
alpha = 0.05

# применяем метод U-критерий Манна-Уитни
results = st.shapiro(group)
p_value = results[1]

# выводим р-значение
print('p-значение: ', p_value)

# формулируем вывод
if p_value < alpha:
    print('Отвергаем нулевую гипотезу: распределение не нормально')
else:
    print('Не получилось отвергнуть нулевую гипотезу, распределение нормально') 
```

Поскольку распределение не нормально и среднее по отточным клиентам меньше среднего по лояльным клиентам - 
применим метод U-критерий Манна-Уитни (st.mannwhitney) с односторонней альтернативой в меньшую сторону (alternative='less')

Основная гипотеза Н0 - средний возраст отточного и лояльного клиента одинаков;                 
Альтернативная гипотеза Н1 - средний возраст отточного клиента меньше среднего возраста лояльного клиента.



```python
# зададим данные для сравнения
group1 = df.query('churn == 1')['age'].astype(int)
group2 = df.query('churn == 0')['age'].astype(int)

# задаем критический уровень статистической значимости
alpha = 0.05

# применяем метод U-критерий Манна-Уитни
results = st.mannwhitneyu(group1, group2, use_continuity=True, alternative='less')

# выводим р-значение
print('p-значение: ', results.pvalue)

# формулируем вывод
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя'
    ) 
```

<div style="border:solid #d9d9d9 2px; padding: 10px">
Можем статистически значимо утверждать, что средний возраст отточного клиента ниже, чем у лояльного клиента.           
Однако в этом случае кроме общей декларации этого дать каки-либо более конкретные рекомендации довольно сложно, поскольку разница в среднем возрасте крайне незначительна.

 ### Проверка гипотезы для параметра оценки заработной платы <a id='step_4.2'></a>

Поскольку распределение не нормально и среднее по отточным клиентам больше среднего по лояльным клиентам - 
применим метод U-критерий Манна-Уитни (st.mannwhitney) с односторонней альтернативой в большую сторону (alternative='greater')

## Шаг 5. Выделение и приоритезация основных сегментов отточных клиентов  <a id='step_5'></a>

### Выделение основных сегментов отточных клиентов <a id='step_5.1'></a>

По результатам проведения исследовательского анализа данных с мы выявили следующее разбиение признаков на интервалы с точки зрения преобладания в этих интервалах отточных клиентов:

1. Кредитный скоринг: от 820 до 920;                           
2. Возраст клиента: от 23 до 35 лет (первый) и от 50 до 60 лет (второй);                          
3. Баланс на счете: от 500 тысяч рублей до максимума (max=119M);                          
4. Количество объектов собственности: от 3 до максимума (max=9);                      
5. Количество используемых продуктов банка: от 3 до максимума (max=5).  

При этом признак оценки заработной платы исключим из дальнейшего рассмотрения по причине того, что его корреляция с практически всеми другими параметрами существенно выше, чем у других параметров. Это вполне объяснимо, поскольку мы имеем дело лишь с оценкой заработной платы, сделанной самим банком на основе как раз других характеристик человека как клиента банка.

Кроме того хорошо бы учесть два бинарных признака, которые мы отмечали как дополнительные.

1. Пол: мужчина (1) - уровень оттока выше, женщина (0) - уровень оттока ниже;                                               
2. Активность пользования услугами банка: активен (1) - уровень оттока выше, не активен (0) - уровень оттока ниже.

При этом признак владения кредитной картой мы не будем включать в рассмотрение в качестве отдельного параметра, поскольку степень корреляции этого признака с перечисленными выше признаками существенно выше, чем у признаков пола и активности (0.2-0.4 против 0.01-0.1).
Оставшиеся в рассмотрении бинарные признаки (пол и активность) хороши еще и тем, что подразумевают возможность несколько разных механизмов работы и меседжей транслируемых этим сегментам (мужчины и женщины имеют разную эмоциональную восприимчивость и им нужны несколько разные меседжи, ну а с активностью тоже все по разному: если маркетинговыми способами активировать неактивного клиента - то его лояльность повысится автоматически, а к ужэ активным нужен другой подход чтобы удержать).

Таким образом мы получаем в рассмотрение 
- 5 основных параметров, один из которых имеет два интересных нам диапазона, остальные по одному, 
- и 2 бинарных параметра.

Нам так-же следует учитывать, что размер результирующих сегментов должен быть "разумен" с точки зрения его позиционирования и возможности банка в дальнейшем работать с ним. В соответстии с разъяснениями заказчика разумный размер сегмента не может превышать 2000 человек.

Сначала составим список находящихся у нас в работе отточных диапазонов по параметрам.

```python
# задаем названия диапазонам
churn_diap_nams =['Кредитный скоринг от 820 до 920 баллов', 
                   'Возраст от 23 до 35 лет', 
                   'Возраст от 50 до 60 лет',
                   'Баланс счета больше 500 тысяч рублей', 
                   'В собственности больше 3 объектов',
                   'В пользовании больше 3 продуктов',
                   'Пол мужской',
                   'Активный клиент'
                 ]


# задаем значения диапазонам
churn_diap_vals = [ df['score'].between(820,920, inclusive='both'),
                    df['age'].between(23,35, inclusive='both'),
                    df['age'].between(50,60, inclusive='both'),
                    df['balance'] >= 500000,
                    df['equity'] >= 3,
                    df['products'] >= 3,
                    df['gender'] == 1,
                    df['activity'] == 1
                  ]
```

Теперь посмотрим размеры и параметры сегментов на основе "полных" интервалов по каждому параметру.

```python
# смотрим размеры "чистых" сегментов по параметрам

for i in range(len(churn_diap_vals)):
    print( 'Размер сегмента', churn_diap_nams[i], ' - ', df.loc[(churn_diap_vals[i]),'churn'].count() )
```

<div style="border:solid #d9d9d9 2px; padding: 10px">
Видим, что 

Хотя есть две категории, размер которых меньше 2 тысяч человек, тем не менее можно констатировать, что в  "чистых" категориях довольно много клиентов и поэтому сегменты хорошо бы уменьшить в размере.

Затем проверим получившиеся сегменты на размер и после сделаем их приоритизацию.

Для каждой комбинации из трех условий, сформированных как мы обозначили выше, посчитаем размер получившегося сегмента, количество отточных клиентов в сегменте и долю отточных клиентов в сегменте и сохраним эту информацию в отдельном датафрейме.

```python
churn_segs = pd.DataFrame(columns = ['cond_1', 'cond_2', 'cond_3', 'seg_size', 'churn_cnt', 'churn_share'])

# задаем циклы по первому, второму и третьему условию
for c1 in range(len(churn_diap_vals)): 
    for c2 in range(len(churn_diap_vals)):
        if c2 <= c1:
            continue
        else:
            for c3 in range(len(churn_diap_vals)):
                if (c3 <= c2) | (c3 <= c1):
                    continue
                else:
                    churn_segs.loc[len(churn_segs)] = [
                                                       churn_diap_nams[c1], 
                                                       churn_diap_nams[c2], 
                                                       churn_diap_nams[c3],

                                                round( df.loc[(churn_diap_vals[c1]) & \
                                                              (churn_diap_vals[c2]) & \
                                                              (churn_diap_vals[c3]), \
                                                              'churn'].count(), 0),
                                                round( df.loc[(churn_diap_vals[c1]) & \
                                                              (churn_diap_vals[c2]) & \
                                                              (churn_diap_vals[c3]), \
                                                              'churn'].sum(), 0),
                                                round( df.loc[(churn_diap_vals[c1]) & \
                                                              (churn_diap_vals[c2]) & \
                                                              (churn_diap_vals[c3]), \
                                                              'churn'].mean(), 2)

                                                          ]
            
# смотрим что получилось
print(churn_segs.shape[0])
display(churn_segs.sort_values('seg_size', ascending=False).head(10))
```
